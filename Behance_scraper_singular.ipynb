```python
#Asignar una variable a response.json() para que sea más fácil acceder
resultado_json = response.json()
```

```python
#Visualizar la información dentro de la clave 'search'
#### COMENTADO PARA AHORRAR ESPACIO ####
#resultado_json['search']
```

```python
#Asignar una variable a resultado_json['search']['content']['projects'] para que sea más fácil acceder
search_results = resultado_json['search']['content']['projects']
```

```python
#Crear listas vacías a las que vamos a añadir nuestros datos
publicación = []
nombre = []
apellido = []
usuario = []
país = []
vistas = []
likes = []
comentarios = []
link = []
```

```python
#Loop que añade los datos obtenidos del website a cada una de las listas

for result in search_results:
    publicación.append(result['name'])    
    nombre.append(result['owners'][0]['first_name'])    
    apellido.append(result['owners'][0]['last_name'])    
    usuario.append(result['owners'][0]['username'])    
    país.append(result['owners'][0]['country'])    
    vistas.append(result['stats']['views'])    
    likes.append(result['stats']['appreciations'])    
    comentarios.append(result['stats']['comments'])    
    link.append(result['url'])
```

```python
#Crear y visualizar un dataframe para estimar si nuestros resultados son correctos
behance_df_singular = pd.DataFrame({'Publicación':publicación, 'Nombre':nombre, 'Apellido':apellido, 'Usuario':usuario, 'País':país, 'Vistas':vistas, 'Likes':likes, 'Comentarios':comentarios, 'Link':link})
behance_df_singular
```
