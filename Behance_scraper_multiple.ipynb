```python
#Creamos nuestras listas vacías como segundo paso porque lo que viene después de esto es horripilante
publicación = []
nombre = []
apellido = []
usuario = []
país = []
vistas = []
likes = []
comentarios = []
link = []
```

```python
#Loop que modifica el argumento 'ordinal' para darnos la cantidad de resultados que buscamos

for i in range(0,480,48):
    params = (
        ('content', 'projects'),
        ('field', 'architecture'),
        ('ordinal', str(i)),
    )
    
    #Todas nuestras santas variables ahora tienen que estar dentro del loop
    response = requests.get('https://www.behance.net/search', headers=headers, params=params)
    resultados_json = response.json()
    search_results = resultados_json['search']['content']['projects']
    
    #Tenemos un loop dentro de un loop por favor que se acabe este sufrimiento
    #Loop que añade los datos obtenidos del website a cada una de las listas
    for result in search_results:
        publicación.append(result['name'])    
        nombre.append(result['owners'][0]['first_name'])    
        apellido.append(result['owners'][0]['last_name'])    
        usuario.append(result['owners'][0]['username'])    
        país.append(result['owners'][0]['country'])    
        vistas.append(result['stats']['views'])    
        likes.append(result['stats']['appreciations'])    
        comentarios.append(result['stats']['comments'])    
        link.append(result['url'])
```

```python
#Crear y visualizar un dataframe para estimar si nuestros resultados son correctos
behance_df_multiple = pd.DataFrame({'Publicación':publicación, 'Nombre':nombre, 'Apellido':apellido, 'Usuario':usuario, 'País':país, 'Vistas':vistas, 'Likes':likes, 'Comentarios':comentarios, 'Link':link})
behance_df_multiple
```
